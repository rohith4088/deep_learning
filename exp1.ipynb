```python
x = df.drop('Salary' , axis = True)
y = df['Salary']
```

```python
from sklearn.model_selection import train_test_split
xtrain , xtest , ytrain , ytest = train_test_split(x , y  ,test_size = 0.2 , random_state = 43)
```

```python
xtrain.shape , ytrain.shape , xtest.shape , ytest.shape
```

```python
sns.pairplot(df)
```

```python
plt.plot(xtrain , ytrain ,'b.' ,label = 'train data')
plt.xlabel('xtrain')
plt.xlabel('ytrain')
plt.legend()
plt.show()
```

```python
#trainig data
plt.plot(xtest , ytest , 'r.' , label ='test data')
plt.xlabel('xtest')
plt.xlabel('ytest')
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(xtrain,ytrain)
```

```python
ypred = regressor.predict(xtest).round(1)
```

```python
calc = pd.DataFrame(np.c_[ytest , ypred],columns = ['original' , 'predicted'])
calc
```

```python
print("training accuracy-->", regressor.score(xtrain ,ytrain ))
print("testing accuracy-->",regressor.score(xtest,ytest))
```
